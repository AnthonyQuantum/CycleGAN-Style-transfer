##**Visualization**

Если посмотреть на результаты, то на них изменился цвет неба, а грунт стал более песчаным и тоже изменил цвет.

```
import matplotlib.pyplot as plt
import numpy as np

def show_imgs(imgs, titles):
  n_images = len(imgs)
  fig = plt.figure()
  for n, (image, title) in enumerate(zip(imgs, titles)):
      a = fig.add_subplot(1, np.ceil(n_images/1.), n + 1)
      a.set_xticks([])
      a.set_yticks([])
      if image.ndim == 2:
          plt.gray()
      plt.imshow(image)
      a.set_title(title)
  fig.set_size_inches(np.array([3., 1.]) * n_images)
  plt.show()

IMAGES_DIR = './results/earth2mars_v1/test_latest/images/'

real_img_names = sorted([img for img in os.listdir(IMAGES_DIR) if img.endswith('_rec_A.png')])
real_imgs = [plt.imread(IMAGES_DIR + name) for name in real_img_names]
show_imgs(real_imgs, real_img_names)

generated_img_names = sorted([img for img in os.listdir(IMAGES_DIR) if img.endswith('_fake_B.png')])
generated_imgs = [plt.imread(IMAGES_DIR + name) for name in generated_img_names]
show_imgs(generated_imgs, generated_img_names)
```
